```python
import pandas as pd
df= pd.read_excel('22_Melbourne_housing.xlsx')
df.head()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.nunique()
```

```python
df.shape
```

```python
col = ['Suburb' , 'Rooms' , 'Type' , 'Method' , 'SellerG' , 'Regionname' , 'Propertycount' ,
       'Distance' , 'CouncilArea' , 'Bedroom2' , 'Bathroom' , 'Car' , 'Landsize' ,
       'BuildingArea' , 'Price']
df = df[col]
df.head()
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
col_fill_zero = ['Propertycount' , 'Distance' , 'Bedroom2' , 'Bathroom' , 'Car']
df[col_fill_zero] = df[col_fill_zero].fillna(0)
df.isna().sum()
```

```python
col_fill_mean = ['Landsize' , 'BuildingArea']
df[col_fill_mean] = df[col_fill_mean].fillna(df[col_fill_mean].mean())
df.isna().sum()
```

```python
df= df.dropna()  #drops all rows with any NaN values from the DataFrame df.
df.isna().sum()
```

```python
df.head()
```

```python
df = pd.get_dummies(df , drop_first = True).astype(int)
df.head()
```

```python
input = df.drop('Price' ,axis=1)
target = df['Price']
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(input , target , test_size = 0.3 , random_state = 2)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X_train , y_train)
```

```python
reg.score(X_test,y_test)
```

```python
reg.score(X_train,y_train)
```

the above train sameple good accuracy but not in testing. implicated overfitting.

```python
from sklearn import linear_model
lasso_reg = linear_model.Lasso(alpha = 50 , max_iter =100 , tol = 0.1)
lasso_reg.fit(X_train , y_train)
```

```python
lasso_reg.score(X_test , y_test)
```

```python
lasso_reg.score(X_train , y_train)
```

```python
from sklearn import linear_model
ridge_reg = linear_model.Ridge(alpha = 50 , max_iter = 100 , tol = 0.1)
ridge_reg.fit(X_train , y_train)
```

```python
ridge_reg.score(X_test , y_test)
```

```python
ridge_reg.score(X_train , y_train)
```
