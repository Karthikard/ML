```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
#dir() to list all the attributes and methods of the iris object
dir(iris)
```

```python
df=pd.DataFrame(iris.data,columns=iris.feature_names)
df.head()
```

```python
df['target'] = iris.target
df.head()
```

```python
iris.target_names
```

```python
df[df.target==1].head()
```

```python
df['flower_name'] = df.target.apply(lambda x: iris.target_names[x])
df.head()
```

```python
#to appear the plot below itself
import matplotlib.pyplot as plt
%matplotlib inline
df0 = df[df.target == 0]
df1 = df[df.target == 1]
df2 = df[df.target == 2]
```

```python
df1.head()
```

```python
plt.xlabel('sepal length')
plt.ylabel('sepal width')
plt.scatter(df0['sepal length (cm)'] , df0['sepal width (cm)'] ,marker='+' , color = 'green')
plt.scatter(df1['sepal length (cm)'] , df1['sepal width (cm)'],marker='+' , color = 'red')
plt.scatter(df2['sepal length (cm)'] , df2['sepal width (cm)'],marker='+' , color = 'blue')
```

```python
plt.xlabel('petal length')
plt.ylabel('petal width')
plt.scatter(df0['petal length (cm)'] , df0['petal width (cm)'] ,marker='+' , color = 'green')
plt.scatter(df1['petal length (cm)'] , df1['petal width (cm)'],marker='+' , color = 'red')
plt.scatter(df2['petal length (cm)'] , df2['petal width (cm)'],marker='+' , color = 'blue')
```

```python
from sklearn.model_selection import train_test_split
```

```python
df.head()
```

```python
input = df.drop(['target' , 'flower_name'] , axis='columns')
input.head()
```

```python
target = df['target']
target.head()
```

```python
X_train ,  X_test , y_train ,y_test=train_test_split(input , target , test_size=0.2)
```

```python
len(X_train)
```

```python
len(y_train)
```

```python
len(X_test)
```

```python
#SVC stands for Support Vector Classifier . It’s part of the SVM family.
from sklearn.svm import SVC
model =SVC()
```

```python
model.fit(X_train , y_train)
```

```python
model.score(X_test,y_test)
```
