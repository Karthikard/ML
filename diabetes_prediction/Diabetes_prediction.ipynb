```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
```

```python
model_params = {
  'dt': {
      'model': DecisionTreeClassifier(),
      'params': {
          'criterion':['gini','entropy']
      }
  },

  'rf': {
      'model': RandomForestClassifier(),
      'params': {
          'n_estimators':[1,5,10]
      }
  },

  'lr': {
      'model': LogisticRegression(solver='liblinear',multi_class='auto'),
      'params': {
          'C':[1,5,10]
      }
  },

  'knn': {
      'model': KNeighborsClassifier(),
      'params': {
          'n_neighbors':[1,5,10]
      }
  },

  'nb': {
      'model': GaussianNB(),
      'params': {

      }
  }
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
scores = []

for model_name , mp in model_params.items():
  clf = GridSearchCV(mp['model'],mp['params'],cv=5,return_train_score=False)
  clf.fit(input,output)
  scores.append({
      'model':model_name,
      'best_score':clf.best_score_,
      'best_params':clf.best_params_
  })
```

```python
scores
```

```python
df = pd.DataFrame(scores,columns=['model','best_score','best_params'])
df
```

```python
# Get the best logistic regression model from the scores list
best_lr_params = next(item for item in scores if item["model"] == "lr")['best_params']
lr_model = LogisticRegression(solver='liblinear', multi_class='auto', **best_lr_params)
lr_model.fit(X_train, Y_train)
print(f"Logistic Regression Test Score: {lr_model.score(X_test, Y_test)}")
```

```python
# Get the best KNN model from the scores list
best_knn_params = next(item for item in scores if item["model"] == "knn")['best_params']
knn_model = KNeighborsClassifier(**best_knn_params)
knn_model.fit(X_train, Y_train)
print(f"KNN Test Score: {knn_model.score(X_test, Y_test)}")
```
