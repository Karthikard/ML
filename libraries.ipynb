Scikit-learn is a powerful and easy-to-use Python library that helps you build machine learning models quickly and efficiently.

To provide tools to train, test, and evaluate machine learning models using just a few lines of code.

from sklearn.datasets import load_iris
This line imports the Iris dataset, which is a small, built-in dataset in Scikit-learn.

from sklearn.model_selection import train_test_split
This function splits your data into two parts:
Training data ,Testing data

from sklearn.neighbors import KNeighborsClassifier
This imports the K-Nearest Neighbors (KNN) algorithm.
KNN is a simple machine learning method used for classification (like predicting flower type).
âœ… You will use this to train your model.

from sklearn.metrics import accuracy_score
This function is used to check how accurate your model is.
It compares your modelâ€™s predictions with the actual answers.

iris.data	The actual data from the Iris dataset (numbers only)
columns=iris.feature_names	Names for each column (like "sepal length", etc.)

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import pandas as pd

iris=load_iris()
df=pd.DataFrame(iris.data,columns=iris.feature_names)
print(df)

df['species'] = iris.target    # This creates a new column in the DataFrame named "species"
                               # iris.target	This is a list of numbers representing flower species: 0 = setosa, 1 = versicolor, 2 = virginica

df.head()
```

```python
#Split data into training and testing (80% train, 20% test)
# X = features(like sepal petal measurement), y = labels(output : type of flowers 0 - setosa , 1-versicolor , 2-virginica)

x=iris.data
y=iris.target

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1) #Whenever you use train_test_split, it randomly shuffles the data before splitting it.
                                                                                 #If you don't set random_state, the split will be different each time you run the code.
print("Training Features")
print(x_train[:5])
print("\nTraining Labels")
print(y_train[:5])

print("\n\n\nTesting Features")
print(x_test[:5])
print("\nTesting Labels")
print(y_test[:5])



```

```python
#Train a model using KNN
model=KNeighborsClassifier(n_neighbors=3)  #When predicting a flower, look at the 3 closest neighbors in the training data and use their majority class (species) to decide.
model.fit(x_train,y_train)  #	Train the model using training features and labels
```

```python
# Predict on test data
y_pred = model.predict(x_test)

print("Predicted Labels")
print(y_pred[:5])
print("\nActual Labels")
print(y_test[:5])

accuracy=accuracy_score(y_test,y_pred)
print("\n\nAccuracy = ", accuracy)
```
