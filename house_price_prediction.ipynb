```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
from google.colab import files
uploaded=files.upload()
```

```python
df=pd.read_excel("houseprice.xlsx")
df.head()
```

```python
plt.scatter(df.area,df.price , color='green' , marker='+')
plt.xlabel("area")
plt.ylabel("price")
```

hene from the above plot it is understandable limear regression model can be used.
fit : training the linear regression model with available data

```python
reg=linear_model.LinearRegression()
reg.fit(df[['area']],df.price)      #the input feature (must be 2D, so double brackets are used)
```

Do prediction

```python
reg.predict([[3300]])
```

```python
plt.plot(df['area'], reg.predict(df[['area']]), label='Hypothesis (regression line)')
plt.scatter(df.area,df.price,color='green',marker='+')
plt.xlabel("area")
plt.ylabel("Price")
plt.legend()


```

```python
reg.coef_   #m value    y=mx+b
```
