```python
import pandas as pd
df=pd.read_excel('carprices.xlsx')
```

```python
df
```

```python
dummies = pd.get_dummies(df['Car Model'] ,dtype='int')
dummies
```

```python
merged = pd.concat([df,dummies],axis='columns')
merged
```

```python
merged = merged.drop(['Car Model','Audi A5'], axis='columns')
merged
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
x=merged.drop(['Sell Price($)'] , axis='columns')
x
```

```python
y=merged['Sell Price($)']
y
```

```python
model.fit(x,y)
```

```python
model.predict([[45000,4,0,0]])
```

```python
model.predict([[69000,6,1,0]])
```

```python
model.predict([[86000,7,1,0]])
```

```python
model.score(x,y)
```
