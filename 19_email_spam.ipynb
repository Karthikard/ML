```python
import pandas as pd
df=pd.read_excel('spam.xlsx')
df.head()
```

```python
df.groupby('Category').describe()
```

```python
#convert catogory into number
dummies = pd.get_dummies(df.Category)
dummies.head()
```

```python
df = pd.concat([df, dummies], axis=1)
df.head()
```

```python
df=df.drop(['Category','ham'],axis='columns')
df.head()
```

```python
df['spam'] = df['spam'].astype(int)

df.head(3)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.Message,df.spam, test_size=0.25)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
v = CountVectorizer()
x_train_count = v.fit_transform(X_train)
x_train_count.toarray()[:3]
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(x_train_count,y_train)
```

```python
emails = [
    'hey meet me as soon as possible in 7th floor' ,
    'get 20% discount on buying 1000 , exclusive for you'
]
emails_count = v.transform(emails)
model.predict(emails_count)
```

```python
X_test_count = v.transform(X_test.astype(str))
model.score(X_test_count , y_test)
```

```python
from sklearn.pipeline import Pipeline
clf = Pipeline([
    ('vectorizer' , CountVectorizer()),
    ('nb' , MultinomialNB())
])
```

```python
clf.fit(X_train , y_train)
```

```python
clf.predict(emails)
```

```python
clf.score(X_test.astype(str), y_test)
```
