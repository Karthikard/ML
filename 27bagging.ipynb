```python
import pandas as pd
df = pd.read_csv('diabetes.csv')
```

```python
df.describe()
```

```python
df.Outcome.value_counts()
```

```python
x = df.drop('Outcome' , axis='columns')
y=df.Outcome
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
x_scaled[:3]
```

```python
# stratify : class proportions in train and test sets are same as in original dataset.
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, stratify=y, random_state=10)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score

scoress = cross_val_score(DecisionTreeClassifier(), x, y, cv=5)
scoress
```

```python
scoress.mean()
```

bagging classifier

```python
from sklearn.ensemble import BaggingClassifier

bag_model = BaggingClassifier(
    estimator=DecisionTreeClassifier(),
    n_estimators=100,
    max_samples=0.8,  #use 80% of samples
    max_features=1.0,
    bootstrap=True,
    bootstrap_features=False,
    oob_score=True,  #Out of bag
    warm_start=False,
)
bag_model.fit(x_train, y_train)
bag_model.oob_score_
```

```python
bag_model.score(x_test, y_test)
```

```python
bag_model = BaggingClassifier(
    estimator=DecisionTreeClassifier(),
    n_estimators=100,
    max_samples=0.8,  #use 80% of samples
    max_features=1.0,
    bootstrap=True,
    bootstrap_features=False,
    oob_score=True,  #Out of bag
    warm_start=False,
)

scores1 = cross_val_score(bag_model, x, y, cv=5)
scores1
```

```python
scores1.mean()
```

```python
from sklearn.ensemble import RandomForestClassifier

scores = cross_val_score(RandomForestClassifier(n_estimators=50), x, y, cv=5)
scores
```
